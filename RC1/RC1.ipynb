```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('mudandoRC.csv')
df.columns
```

```python
df1 = df[df['RC']==0.001034]
df2 = df[df['RC']==0.0022]
df3 = df[df['RC']==0.0047]
df4 = df[df['RC']<0.0005]
df5 = df[df['RC']==0.0033]
print(df1)
print(df2)
print(df3)
print(df4)
print(df5)
```

```python
plt.subplot(1, 2, 1)
plt.plot(df1['freq'],df1['defasagem '],
         df1['freq'],df2['defasagem '],
         df1['freq'],df3['defasagem '],
         df1['freq'],df4['defasagem '],
         df1['freq'],df5['defasagem ']
        )
plt.legend(['a','b','c','d','e'])
plt.xlabel('Frequncia (Hz)')
plt.ylabel('Defasagem ()')
plt.xscale('log')
plt.subplot(1, 2, 2)
plt.plot(df1['freq'],df1['out_vpp'],
         df1['freq'],df2['out_vpp'],
         df1['freq'],df3['out_vpp'],
         df1['freq'],df4['out_vpp'],
         df1['freq'],df5['out_vpp']
        )
plt.legend(['a','b','c','d','e'])
plt.xlabel('Frequncia (Hz)')
plt.ylabel('Tenso de sada pico a pico(Vpp)')
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
plt.savefig('foo.png')
plt.show()
```
